```python
bboard = pd.read_pickle("R&B10yrDF.pkl")
bsongs = bboard.Title.dtype
spotify_songs = spotify_data[:500].track_name
bboard.dtypes
```

```python
def checkIfHot(spot_song, hit_songs):
    if(spot_song in hit_songs):
        return True
    else:
        return False
spotify_data['isHot'] = spotify_data['track_name'].apply(lambda x: checkIfHot(x, bboard.Title))
```
